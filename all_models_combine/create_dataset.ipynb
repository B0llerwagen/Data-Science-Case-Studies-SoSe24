```python
import pandas as pd

labels = pd.read_csv("../labels.csv", sep=";")
```

```python
test = pd.DataFrame(labels[[all([int(rule) == 0 for rule in label[1:]]) for label in labels.values]])
test.describe()
```

```python
label_1 = pd.DataFrame(labels[labels["Rule1"] == 1])
label_2 = pd.DataFrame(labels[labels["Rule2"] == 1])
label_3 = pd.DataFrame(labels[labels["Rule3"] == 1])
label_4 = pd.DataFrame(labels[labels["Rule4"] == 1])
label_5 = pd.DataFrame(labels[labels["Rule5"] == 1]) 
label_6 = pd.DataFrame(labels[labels["Rule6"] == 1])
label_7 = pd.DataFrame(labels[labels["Rule7"] == 1])
label_8 = pd.DataFrame(labels[labels["Rule8"] == 1])

all_lab = pd.concat([label_1, label_3, label_4, label_5, label_6, label_7, label_8])
all_lab.drop_duplicates()

print(all_lab.sum(numeric_only=True))
print(all_lab.describe())
print(all_lab.info())
print(all_lab.head())
```

```python
balanced_labels = pd.concat([test.sample(int(len(all_lab)*0.25), random_state=42), all_lab])
balanced_labels.describe()
```

```python
balanced_labels.to_csv('../all_models_combine/chosen_labels_all.csv', sep=';', index=False)
```
