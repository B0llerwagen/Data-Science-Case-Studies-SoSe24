```python
#load model and testing data
from tensorflow import keras
import pickle

model = keras.models.load_model('../model_julius/model_all_1.h5')

with open('../model_julius/all_test_img.pickle', 'rb') as f:
    X_test, y_test = pickle.load(f)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Ensure X_test, y_test, and predicted_classes are defined and accessible in this cell
X_test = X_test.astype(np.float32)

fig, axes = plt.subplots(5, 1, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    idx = np.random.randint(X_test.shape[0])
    ax.imshow(X_test[idx].astype('uint8'))
    ax.set_title(f"Label: {y_test[idx:idx+1]}, Prediction: {(model.predict(X_test[idx:idx+1].astype(np.float32)))}")
    ax.axis('off')
plt.tight_layout()
plt.show()
```
