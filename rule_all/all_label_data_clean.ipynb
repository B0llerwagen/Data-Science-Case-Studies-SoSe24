```python
#load data
import pandas as pd

labels = pd.read_csv('../labels.csv', header=0, sep=';')
#print(labels.head())
labels.info()
```

```python
#take all images with at least one problem
import pandas as pd
label_1 = pd.DataFrame(labels[[any([int(rule) == 1 for rule in label[1:]]) for label in labels.values]])
print(label_1.info())

label_0 = pd.DataFrame(labels[[all([int(rule) == 0 for rule in label[1:]]) for label in labels.values]])
print(label_0.info())

balanced_labels = pd.concat([label_0, label_1])
```
